### add additional features

```python
reg_season_detail['Wfgpct'] = reg_season_detail['Wfgm']*1.0/reg_season_detail['Wfga']
reg_season_detail['Wfgpct3'] = reg_season_detail['Wfgm3']*1.0/reg_season_detail['Wfga3']
reg_season_detail['Wftpct'] = reg_season_detail['Wftm']*1.0/reg_season_detail['Wfta']
reg_season_detail['Wrbpct'] = (reg_season_detail['Wor']+reg_season_detail['Wdr'])/(reg_season_detail['Wor']+reg_season_detail['Wdr']+reg_season_detail['Lor']+reg_season_detail['Ldr'])
reg_season_detail['Wrbdiff'] = (reg_season_detail['Wor']+reg_season_detail['Wdr']) - (reg_season_detail['Lor']+reg_season_detail['Ldr'])
reg_season_detail['Wrb'] = reg_season_detail['Wor']+reg_season_detail['Wdr']

reg_season_detail['Lfgpct'] = reg_season_detail['Lfgm']*1.0/reg_season_detail['Lfga']
reg_season_detail['Lfgpct3'] = reg_season_detail['Lfgm3']*1.0/reg_season_detail['Lfga3']
reg_season_detail['Lftpct'] = reg_season_detail['Lftm']*1.0/reg_season_detail['Lfta']
reg_season_detail['Lrbpct'] = (reg_season_detail['Lor']+reg_season_detail['Ldr'])/(reg_season_detail['Wor']+reg_season_detail['Wdr']+reg_season_detail['Lor']+reg_season_detail['Ldr'])
reg_season_detail['Lrbdiff'] = (reg_season_detail['Lor']+reg_season_detail['Ldr']) - (reg_season_detail['Wor']+reg_season_detail['Wdr'])
reg_season_detail['Lrb'] = reg_season_detail['Lor']+reg_season_detail['Ldr']

reg_season_detail['Wdiff'] = reg_season_detail['Wscore']-reg_season_detail['Lscore']
reg_season_detail['Ldiff'] = reg_season_detail['Lscore']-reg_season_detail['Wscore']
```
